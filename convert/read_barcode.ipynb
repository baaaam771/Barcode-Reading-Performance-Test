```python
import os
import json

# 이미지 파일들이 들어있는 폴더 경로
image_folder = 'crop100'

# 이미지 파일이 저장된 경로를 순회
for filename in os.listdir(image_folder):
    if filename.endswith(('.jpg', '.png', '.jpeg')):  # 이미지 파일만 선택
        # 파일명에서 확장자를 제거하여 바코드 번호 추출
        barcode_number = os.path.splitext(filename)[0]

        # JSON 형식으로 ground truth 생성
        ground_truth = [
            {
                "attrib": {"Type": "EAN13"},  # EAN13 바코드 유형
                "text": barcode_number,       # 파일명에서 추출한 바코드 번호
                "value_attrib": {}
            }
        ]

        # JSON 파일명 생성 (이미지 파일명과 동일한 이름을 사용, 확장자는 .txt로 변경)
        json_filename = f"{barcode_number}.txt"
        json_path = os.path.join(image_folder, json_filename)

        # JSON 파일로 저장
        with open(json_path, 'w') as json_file:
            json.dump(ground_truth, json_file, indent=4)

        print(f"Created {json_filename} for {filename}")
```

```python
import os
import json

# 이미지 파일들이 들어있는 폴더 경로
image_folder = 'crop100'

# 이미지 파일이 저장된 경로를 순회
for filename in os.listdir(image_folder):
    if filename.endswith(('.jpg', '.png', '.jpeg')):  # 이미지 파일만 선택
        # 파일명에서 확장자를 제거하고 '_cropped'를 제거하여 바코드 번호 추출
        barcode_number = os.path.splitext(filename)[0].replace('_cropped', '')

        # JSON 형식으로 ground truth 생성
        ground_truth = [
            {
                "attrib": {"Type": "EAN13"},  # EAN13 바코드 유형
                "text": barcode_number,       # 파일명에서 추출한 바코드 번호
                "value_attrib": {}
            }
        ]

        # JSON 파일명 생성 (이미지 파일명과 동일한 이름을 사용, 확장자는 .txt로 변경)
        json_filename = f"{barcode_number}.txt"
        json_path = os.path.join(image_folder, json_filename)

        # JSON 파일로 저장
        with open(json_path, 'w') as json_file:
            json.dump(ground_truth, json_file, indent=4)

        print(f"Created {json_filename} for {filename}")
```
